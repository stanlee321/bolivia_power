```python

import plotly.plotly as py
from plotly.graph_objs import *

import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns 
from mpl_toolkits.basemap import Basemap

#import semanticnet as sn

%matplotlib inline
```

```python
dirname = "/notebooks/Power/data/proyect/data_bolivian_power/full/"

# Load every image file in the provided directory
filenames = [os.path.join(dirname, fname)
             for fname in os.listdir(dirname)]

filenames = filenames[:]

```

```python
# Cheking the names of the filenames into the filenames list... as you can see the lenght of the list is 581
for i in range(len(filenames)):
    print(i,filenames[i])
```

```python
# Load all the files as dataframes into a list
files_csv = []

def create_df():
    global files_csv
    for i in range(len(filenames)):
        data_frame = pd.read_csv(filenames[i])
        files_csv.append(data_frame)
        print(i,files_csv[-1])
create_df()        
```

```python
# Cleaning the zero values of the dataframes into the files_csv file
clean_list = []
for k in range(len(files_csv)):
    if len(files_csv[k]) != 0:
        clean_list.append(files_csv[k])
        print(k,len(files_csv[k]))
```

```python
# Cheking the data inside of the clean_list.. the lenght of  the list of datafranes to work is just of 95 
for i in range(len(clean_list)):
    print(i,clean_list[i])
```

```python
print(len(clean_list))
```

```python
# Lets working with the first value of the list

df_0 = clean_list[0]
df_0 = df_0.drop('Unnamed: 0', 1)

df_0["fromBus"] = df_0["fromBus"].astype(np.int) 
df_0["toBus"]   = df_0["toBus"].astype(np.int) 

df_0.head()
```

```python
# As you can see this is a line
```

```python
# Load the full df_nodes data

df_nodes = pd.read_csv("/notebooks/Power/data/nodes_clean.csv") # Load nodes_info

df_nodes.head()
```

```python
# Code for check the distance > 2.5 km in our df_0
mask = df_0["distance[km]"] > 2.5
long_lines = df_0[mask]
#Output
long_lines[["fromBus", "toBus", "distance[km]"]]
```

```python
# There is no exist such case
```

```python
# appending the fromBus and toBUs in one dataFrame for purpouse of filtering into the df_nodes
M = df_0["fromBus"].append(df_0["toBus"])
M.shape
```

```python
# Droping duplicates and putting all together into a list called L

L = M.drop_duplicates().tolist()
#print(L)
```

```python
# Filterning the ids that apear into the df_nodes  from df_0  and reseting index to 0 from the value of the original df
df_nodes_df_0 = df_nodes[df_nodes.id.isin(L)].reset_index()
df_nodes_df_0[["id","lat","lon"]].head()
```

```python
# Calculating the max and min values of lat and long respectively

maxi_lat = max(df_nodes_df_0 ["lat"])
mini_lat = min(df_nodes_df_0 ["lat"])

maxi_lon = max(df_nodes_df_0 ["lon"])
mini_lon = min(df_nodes_df_0 ["lon"])
```

```python
df_nodes_df_0.shape # just 14 values to plot
```

```python
# Ploting the values of the nodes and his edges "the line" into Bolivan map
import networkx as nx
# Sample it down to only the Bolivian region
#lon_min, lon_max = -70.752, -23.483-33
#lat_min, lat_max = -56.382+33, -9.449

# Sample it down the values of lat and long of the line
size_exp = 1.001

lon_min, lon_max = mini_lon, maxi_lon 
lat_min, lat_max = mini_lat, maxi_lat 

idx_bol = (df_nodes_df_0["lon"]>lon_min) &\
            (df_nodes_df_0["lon"]<lon_max) &\
            (df_nodes_df_0["lat"]>lat_min) &\
            (df_nodes_df_0["lat"]<lat_max)

df_nodes_bol_df_0 = df_nodes_df_0[idx_bol]


# Mercator of Bolivia
plt.figure(4, figsize=(24,24)) # Size of the image

m = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max,
             lat_ts=35,
             resolution='i')

#m.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
#m.drawmapboundary(fill_color='#000000')                # black background
m.drawcountries(linewidth=0.4, color="black")              # thin white line for country borders
m.drawstates(linewidth=0.4, color="black")

##################################################################

# Plot the data
#mxy = m(df_nodes_bol["lon"].tolist(), df_nodes_bol["lat"].tolist())
#m.scatter(mxy[0], mxy[1], s=30, c="#1292db", lw=0, alpha=1, zorder=8)


# load geographic coordinate 

node = df_nodes_df_0["id"].values.tolist()    # clear spaces
lat = df_nodes_df_0["lat"].values.tolist()
lon = df_nodes_df_0["lon"].values.tolist()


# Create Graph over the points going to be plot
G=nx.Graph()


# Adding the edges of the nodes "lines"
for i in range(df_0.shape[0]):
    G.add_edge(df_0["fromBus"][i],df_0["toBus"][i])
      

# Dict over which the points are goint to be associated        
pos = {} 
for i in range(0, len(node)):

    pos[node[i]] = m(lon[i], lat[i])

# Draw the full Graph 
nx.draw_networkx(G, pos,node_size=100,node_color='#1292db',edge_color = 'red')


plt.title("Bol view of events")
plt.show()
```
